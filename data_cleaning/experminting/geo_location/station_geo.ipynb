```python
import json
import matplotlib.pyplot as plt
```

```python
station_geo_location = {}

with open('centers_normalized.json', 'r') as data_file:
    data_loaded = json.load(data_file)
    data_keys = list(data_loaded.keys())
    for each_key in data_keys:
        formated = each_key.split('-')
        if formated[0] not in list(station_geo_location.keys()):
            station_geo_location[formated[0]] = []
        station_geo_location[formated[0]].append(data_loaded[each_key])
            
for each_station in list(station_geo_location.keys()):
    plt.scatter([i[0] for i in station_geo_location[each_station]], [i[1] for i in station_geo_location[each_station]], label=each_station)

plt.legend()
plt.show()
```
