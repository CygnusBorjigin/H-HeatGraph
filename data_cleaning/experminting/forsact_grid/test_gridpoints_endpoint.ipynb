```python
import json
import requests
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
possible_point = [(a, b) for a in range(200) for b in range(200)]
successful_point = []
failed_points = []

for each_point in tqdm(possible_point[:1000]):
    respond = requests.get(f"https://api.weather.gov/gridpoints/AKQ/{each_point[0]},{each_point[1]}")
    if respond.status_code == 200:
        successful_point.append(each_point)
    else:
        failed_points.append(each_point)
```

```python
weather_stations = []

for i in range(1,9):
    with open(f'weather_station_new_england_{i}.json', 'r') as data_file:
        data_loaded = json.load(data_file)
        data_keys = list(data_loaded.keys())
        for each_key in data_keys:
            if each_key not in weather_stations and each_key != "null":
                weather_stations.append(each_key)

okx_coordinates = []
box_coordinates = []
aly_coordinates = []
gyx_coordinates = []
btv_coordinates = []
car_coordinates = []

for i in range(1, 9):
    with open(f'weather_station_new_england_{i}.json', 'r') as data_file:
        data_loaded = json.load(data_file)
        data_keys = list(data_loaded.keys())
        if "OKX" in data_keys:
            okx_coordinates.extend(data_loaded["OKX"])
        if "BOX" in data_keys:
            box_coordinates.extend(data_loaded["BOX"])
        if "ALY" in data_keys:
            aly_coordinates.extend(data_loaded["ALY"])
        if "GYX" in data_keys:
            gyx_coordinates.extend(data_loaded["GYX"])
        if "BTV" in data_keys:
            btv_coordinates.extend(data_loaded["BTV"])
        if "CAR" in data_keys:
            car_coordinates.extend(data_loaded["CAR"])
```

```python
plt.scatter([i[0] for i in okx_coordinates], [i[1] for i in okx_coordinates])
plt.title("OKX")
plt.legend()
plt.show()

plt.scatter([i[0] for i in box_coordinates], [i[1] for i in box_coordinates])
plt.title("BOX")
plt.legend()
plt.show()

plt.scatter([i[0] for i in aly_coordinates], [i[1] for i in aly_coordinates])
plt.title("ALY")
plt.legend()
plt.show()

plt.scatter([i[0] for i in gyx_coordinates], [i[1] for i in gyx_coordinates])
plt.title("GYX")
plt.legend()
plt.show()

plt.scatter([i[0] for i in btv_coordinates], [i[1] for i in btv_coordinates])
plt.title("BTV")
plt.legend()
plt.show()

plt.scatter([i[0] for i in car_coordinates], [i[1] for i in car_coordinates])
plt.title("CAR")
plt.legend()
plt.show()
```
